```python
import random
import time
import matplotlib.pyplot as plt

# Linear search implementation
def linear_search(arr, key):
    for i in range(len(arr)):
        if arr[i] == key:
            return i
    return -1

# Generate random list
n = 10000
data = [random.randint(0, 100000) for _ in range(n)]

# Define keys for each case
best_case_key = data[0]
average_case_key = data[n // 2]
worst_case_key = -1  # Not in list

# Timing function
def measure_time(arr, key):
    start = time.time()
    linear_search(arr, key)
    end = time.time()
    return (end - start) * 1000  # milliseconds

# Measure time for each case
best_time = measure_time(data, best_case_key)
average_time = measure_time(data, average_case_key)
worst_time = measure_time(data, worst_case_key)

# Print execution times
print(f"Best Case Time: {best_time:.4f} ms")
print(f"Average Case Time: {average_time:.4f} ms")
print(f"Worst Case Time: {worst_time:.4f} ms")

# Plot results
cases = ['Best Case', 'Average Case', 'Worst Case']
times = [best_time, average_time, worst_time]

plt.figure(figsize=(8, 5))
plt.bar(cases, times, color=['green', 'orange', 'red'])
plt.title('Linear Search Time Analysis (n=10000)')
plt.xlabel('Case Type')
plt.ylabel('Time (ms)')
plt.grid(True)
plt.show()

# Theoretical analysis
print("\nTheoretical Time Complexities:")
print("Best Case: O(1) - Key found at the beginning.")
print("Average Case: O(n/2) ≈ O(n) - Key found near the middle.")
print("Worst Case: O(n) - Key not present, all elements checked.")

print("\nObservation:")
print("As input size increases, time increases linearly in the worst case.")
```
